# pandas:parallel coordinate

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize']=[5,5]
pd.plotting.parallel_coordinates(iris_df, 'target')
#ข้อจำกัดอันนี้ ดูดาต้าเยอะๆไม่ค่อยดี ถ้าคอลัมเยอะจะไม่ดี
```

# pandas: scatter matrix

```python
pd.plotting.scatter_matrix(iris_df);
#เพื่อดูความสัมพันกับตัวเอง ว่ามีค่าไหนบ้างที่อยุ่ในแต่ละคอลัม
#ดูเทรนของดาต้า
```

# color

```python
#คำนึงถึงคนตาบอดสี บางคุ่ีมองไม่เห้น และ สีที่มีความเกี่ยวข้องกับข้อมูล
matplotlib.rcParams['figure.figsize']=[10,45]
plt.imshow(iris_df.iloc[:,:], interpolation='nearest',cmap='twilight')
```

```python
matplotlib.rcParams['figure.figsize']=[5,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=('red','green','blue')) #ใช้ชื่อสีในการกำหนดสี
```

```python
matplotlib.rcParams['figure.figsize']=[5,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=('c','m','y')) #ใช้ตัวย่อของสี c m y k r g b w
```

```python
matplotlib.rcParams['figure.figsize']=[5,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=[[1,0,0],[0.5,0.5,0.5],[0,1,1]]) #ใช้ค่าสี R G B
```

```python
matplotlib.rcParams['figure.figsize']=[5,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=['#FC9EFF','#B0DE3C','#3CD6DE']); #ใช้ HTML color
```
